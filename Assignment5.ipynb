```python
# Exercise 6.1
# Consider the dataset from `data_banknote_authentication.csv`.
# 1) Read data into a pandas dataframe. 


import pandas as pd

df = pd.read_csv('data_banknote_authentication.csv')

print(df.head())
```

```python
# 2. Pick the column named "class" as target variable `y` and all other columns as feature variables `X`.
X = df.drop(columns=['class'])

y = df['class']

print("Features shape (X):", X.shape)
print("Target shape (y):", y.shape)

print("\nFirst 5 rows of features (X):")
print(X.head())
```

```python
# 3. Split the data into training and testing sets with 80/20 ratio and `random_state=20`.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=20)

print(
"X_train shape:", X_train.shape, "| y_train shape:", y_train.shape)
print("X_test shape: ", X_test.shape, " | y_test shape: ", y_test.shape)
```

```python
# 4. Use support vector classifier with linear kernel to fit to the training data.
from sklearn.svm import SVC

linear_svm = SVC(kernel='linear', random_state=20)

linear_svm.fit(X_train, y_train)

print("Linear SVM model trained successfully!")
```

```python
# 5. Predict on the testing data and compute the confusion matrix and classification report.

from sklearn.metrics import confusion_matrix, classification_report

y_pred = linear_svm.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print("--- Confusion Matrix ---")
print(cm)

report = classification_report(y_test, y_pred)
print("\n--- Classification Report ---")
print(report)
```

```python
# 6. Repeat steps 3 and 4 for the radial basis function kernel.

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=20)

rbf_svm = SVC(kernel='rbf', random_state=20)

rbf_svm.fit(X_train, y_train)

print("RBF SVM model trained successfully!")
```

```python
# 7. Compare the two SVM models in your own words.
The differences between Linear and RBF SVMs
 Decision Boundary:
Linear Kernel: It will draw a straight line that separates the classes, which means that the data can be separated by a linear rule.
RBF Kernel: It produces a curved and flexible decision boundary. It first transforms the data into a higher-dimensional space where the non-linear patterns can be separated.
Performance on this Dataset
Linear SVM: It can deliver high-quality results since the banknote features easily separate along a straight boundary.
RBF SVM: It can not only reach the level of the linear model but often exceeds it, due to its curved boundary that perfectly includes the few edge cases that are hard to separate.
Complexity & Risk
Linear SVM: Less complicated, extremely quick to train, easy to interpret, and has a very low chance of overfitting.
RBF SVM: Very flexible and powerful, but slower to operate and more prone to overfitting if its parameters (C and gamma) are not carefully regulated.
```

```python
# Exercise 6.2
# This exercise is related to exercise 5.2 of the previous week. Consider the data from CSV file `weight-height.csv`.
# Read data into a pandas dataframe.

import pandas as pd

df = pd.read_csv('weight-height (1).csv')

print(df.head())
```

```python
# 2. Pick the target variable `y` as weight in kilograms, and the feature variable `X` as height in centimeters.

df['Height_cm'] = df['Height'] * 2.54


df['Weight_kg'] = df['Weight'] * 0.45359237

X = df[['Height_cm']] 
y = df['Weight_kg']

print("First 5 rows of X (Height in cm):")
print(X.head())

print("\nFirst 5 rows of y (Weight in kg):")
print(y.head())
```

```python
# 3. Split the data into training and testing sets with 80/20 ratio.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print("X_train shape:", X_train.shape, "| y_train shape:", y_train.shape)
print("X_test shape: ", X_test.shape, " | y_test shape: ", y_test.shape)
```

```python
# 4. Scale the training and testing data using normalization and standardization.

from sklearn.preprocessing import MinMaxScaler, StandardScaler

min_max_scaler = MinMaxScaler()

X_train_norm = min_max_scaler.fit_transform(X_train)
X_test_norm = min_max_scaler.transform(X_test)


std_scaler = StandardScaler()

X_train_std = std_scaler.fit_transform(X_train)
X_test_std = std_scaler.transform(X_test)


print("Normalized Training - Min:", X_train_norm.min(), "Max:", X_train_norm.max())
print("Standardized Training - Mean:", round(X_train_std.mean()), "Std:", X_train_std.std())
```

```python
# 5. Fit a KNN regression model with `k=5` to the training data without scaling, predict on unscaled testing data and compute the $R^2$ value.
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

knn_regressor = KNeighborsRegressor(n_neighbors=5)

knn_regressor.fit(X_train, y_train)

y_pred_unscaled = knn_regressor.predict(X_test)

r2_unscaled = r2_score(y_test, y_pred_unscaled)

print(f"R² value on unscaled data: {r2_unscaled:.4f}")
```

```python
# 6. Repeat step 4 for normalized data.
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

knn_regressor_norm = KNeighborsRegressor(n_neighbors=5)

knn_regressor_norm.fit(X_train_norm, y_train)

y_pred_norm = knn_regressor_norm.predict(X_test_norm)

r2_norm = r2_score(y_test, y_pred_norm)

print(f"R² value on normalized data: {r2_norm:.4f}")
```

```python
# 7. Repeat step 4 for standardize data.

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score

knn_regressor_std = KNeighborsRegressor(n_neighbors=5)

knn_regressor_std.fit(X_train_std, y_train)

y_pred_std = knn_regressor_std.predict(X_test_std)

r2_std = r2_score(y_test, y_pred_std)

print(f"R² value on standardized data: {r2_std:.4f}")
```

```python
Since there is only one feature (Height) in the data set, the R^2 values in the three KNN models are same. So, even when you normalize or standardize the data by scaling, you are basically just shrinking or stretching the data points along the axis; Yet, their relative positioning, order, and proximity rankings remain unchanged. Because K-Nearest Neighbors algorithm just relies on relative distances to select the K nearest ones e.g. 5 neighbors, it always picks the same people to be neighbors for a point of test whether or not the data is scaled so predictions and resulting R^2 scores are ultimately the same. Generally, scaling affects performance metric when a model is trained with multiple features differing in units of measurement. Besides, it prevents features with very large numerical values from dominating others during distance computations.
```
